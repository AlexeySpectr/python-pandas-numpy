```python
dd = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve',
             'Frank', 'Grace', 'Hannah', 'Ivan', 'Jack',
             'Kathy', 'Liam', 'Mona', 'Nina', 'Oscar', 'Paul'],
    'age': [25, 30, 22, 35, 28,
            40, 23, 31, 29, 34,
            27, 26, 32, 24, 33, 38]
})
result = dd.loc[[5, 10, 15], ['name', 'age']]
result
```

```python
d1 = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve',
             'Frank', 'Grace', 'Hannah', 'Ivan', 'Jack',
             'Kathy', 'Liam', 'Mona', 'Nina', 'Oscar', 'Paul'],
    'age': [25, 30, 22, 35, 28,
            40, 23, 31, 29, 34,
            27, 26, 32, 24, 33, 38],
    'city': ['NY', 'LA', 'NY', 'LA', 'NY',
             'LA', 'NY', 'LA', 'NY', 'LA',
             'NY', 'LA', 'NY', 'LA', 'NY', 'LA']
})

# Выбираем только строки с определенными именами
sub = d1[d1['name'].isin(['Frank', 'Kathy', 'Paul', 'Alice'])]

# Группируем по городу и считаем средний возраст
resul = sub.groupby('city')['age'].agg(['mean', 'max', 'min']).reset_index()
resul
```

```python
students = pd.DataFrame({
    'student_id': [100, 101, 102, 103, 101],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'age': [20, 21, 19, 22, 21],
    'grade': ['A', 'B', 'A', 'C', 'B']  # Дополнительная колонка
})

# Фильтрация по student_id == 101 и оставляем только name и age
students = students[students['student_id'] == 101]
students = students[['name', 'age']]
students
```

```python
products = pd.DataFrame({
    'product': ['Apple', 'Banana', 'Cherry', 'Date'],
    'quantity': [10, np.nan, 5, np.nan]
})
products['quantity'] = products['quantity'].where(products['quantity'].notna(), 0)
products
```

```python
weather = pd.DataFrame({
    'month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar'],
    'city': ['NY', 'LA', 'NY', 'LA', 'NY', 'LA'],
    'temperature': [30, 60, 32, 62, 45, 65]
})

dg = weather.pivot(index='month', columns='city', values='temperature')
dg
```
